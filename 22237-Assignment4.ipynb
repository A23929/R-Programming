```r
# Load the mtcars dataset
data(mtcars)
head(mtcars)
```

### BAR GRAPH

```r
# Create a table of the number of cars with number of cylinders each
cyl_table <- table(mtcars$cyl)

# Create a bar chart
barplot(cyl_table, main = "Number of Cars by Cylinder Count", xlab = "Number of Cylinders", ylab = "Number of Cars",col='steelblue')
 
```

### HISTOGRAM

```r
# Create a histogram of the car weights
hist(mtcars$wt, main = "Distribution of Car Weights", xlab = "Weight", ylab = "Frequency", col='steelblue')
```

### SCATTERPLOT

```r
# Create a scatterplot of mpg by weight
plot(mtcars$wt, mtcars$mpg, main = "Miles Per Gallon by Car Weight", xlab = "Weight", ylab = "Miles per Gallon")
```

### PIE CHART

```r
# Create a table of the number of cars with each number of gears
gear_table <- table(mtcars$gear)

# Create a pie chart
pie(gear_table, main = "Number of Cars by Gear Count", col = c("brown", "purple", "blue"), labels = c(""))
```

### BOX PLOT

```r
boxplot(hp ~ cyl,mtcars,
        main='Distribution of mpg values',
        ylab='mpg',
        col='steelblue',
        border='black')
```
